```python
def reveal(data, i):
  tmp_s=[i-1]
  s=i-1
  while s<=311:
    s+=31
    tmp_s.append(s)

  print('tmp_s', tmp_s)
  tmp_ = np.zeros((data.shape[0], 0))
  for jdx in tmp_s:
    tmp_ = np.concatenate((tmp_, data[:, jdx:jdx+1]), axis=1)
  print('tmp_.shape', tmp_.shape)
  return tmp_
```
